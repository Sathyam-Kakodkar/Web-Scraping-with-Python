```python
# import the libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Note: The base url and the number of pages on the website keeps changing.

```python
base_url = "https://www.screener.in/screens/214283/companies-creating-new-high/?page="
```

```python
# Specify the number of pages you want to loop through
num_pages = 20
```

```python
for page in range(1, num_pages + 1):
    url = base_url + str(page)

    # Make a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the table with the class 'data-table'
        table = soup.find('table', {'class': 'data-table'})

        # Extract the table headers
        headers = [th.text.strip() for th in table.find_all('th')]

        # Use a set to keep track of unique elements while preserving order
        seen = set()

        # Create a new list with unique elements in the original order
        headers_1 = [x for x in headers if x not in seen and not seen.add(x)]

        # Removing the whitespaces and unwanted items from the string
        headers_2 = []
        for i in headers_1:
            x = ' '.join(i.split())
            if x not in headers_2:
                headers_2.append(x)

        # Extract the data rows
        if page == 1:
            rows = []
            for row in table.find_all('tr')[1:]:
                rows.append([td.text.strip() for td in row.find_all('td')])
        else:
           for row in table.find_all('tr')[1:]:
                rows.append([td.text.strip() for td in row.find_all('td')])




        # Create a Pandas DataFrame
        df = pd.DataFrame(rows, columns=headers_2)


    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
df.to_csv('data.csv', index=False)
```
